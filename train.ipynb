```python
config_file_path = "COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"
checkpoint_url = "COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"
output_dir = "/content/drive/MyDrive/Colab Notebooks_sirisha/instance_segmentation_detectron2"
num_classes = 8

device = "cuda"
```

```python
train_dataset_name = "liquid_train"
train_images_path = "/content/drive/MyDrive/Colab Notebooks_sirisha/Spill Detection.v6i.coco-segmentation_detectron2/train"
train_json_annotation_path = "/content/drive/MyDrive/Colab Notebooks_sirisha/Spill Detection.v6i.coco-segmentation_detectron2/train_annotations.coco.json"

test_dataset_name = "liquid_test"
test_images_path = "/content/drive/MyDrive/Col/content/drive/MyDrive/Colab Notebooks_sirisha/Spill Detection.v6i.coco-segmentation_detectron2/test"
test_json_annotation_path = "/content/drive/MyDrive/Colab Notebooks_sirisha/Spill Detection.v6i.coco-segmentation_detectron2/test_annotations.coco.json"

cfg_save_path = "/content/drive/MyDrive/Colab Notebooks_sirisha/Spill Detection.v6i.coco-segmentation_detectron2/IS_cfg.pickle"
```

```python
class_names = ["Battery Acid", "Bleach", "Coolant", "Fuel", "Hydraulic Fuel",
               "Hydrochloric Acid", "Motor Oil", "VX"]

register_coco_instances(name = train_dataset_name, metadata={"thing_classes": class_names},
json_file = train_json_annotation_path, image_root = train_images_path)

register_coco_instances(name = test_dataset_name, metadata={"thing_classes": class_names},
json_file = test_json_annotation_path, image_root = test_images_path)
```

```python
from utils import get_train_cfg

def main():
  cfg = get_train_cfg(config_file_path, checkpoint_url, train_dataset_name, test_dataset_name, num_classes, device, output_dir)

  with open(cfg_save_path, "wb") as f:
    pickle.dump(cfg, f, protocol=pickle.HIGHEST_PROTOCOL)

  os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)
  trainer = DefaultTrainer(cfg)
  trainer.resume_or_load(resume=False)
  trainer.train()

if __name__ == '__main__':
  main()
```
